# Example 14 - Summarize results

In this example, we'll summarize results, for example, comparison between HC and CNN features. 

## Step 1. Validation results

```python
import pickle
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
sns.set(font_scale=1.3)
sns.set_style("whitegrid", {
    "ytick.major.size": 0.1,
    "ytick.minor.size": 0.05,
    "grid.linestyle": '--',
})
```

```python
pkl_dir = "pkl/"
```

```python
res_df_only_HC_RF = pd.read_pickle(pkl_dir + "res_df_only_HC_RF.pkl")
res_df_only_HC_LR = pd.read_pickle(pkl_dir + "res_df_only_HC_LR.pkl")
res_df_only_HC_SVM = pd.read_pickle(pkl_dir + "res_df_only_HC_SVM.pkl")
res_df_only_ResNet = pd.read_pickle(pkl_dir + "res_df_only_CNN_ResNet.pkl")
res_df_only_VGG = pd.read_pickle(pkl_dir + "res_df_only_CNN_VGG.pkl")
res_df_only_MobileNet = pd.read_pickle(pkl_dir + "res_df_only_CNN_MobileNet.pkl")
res_df_cotraining_HC = pd.read_pickle(pkl_dir + "res_df_cotraining_HC.pkl")
res_df_cotraining_CNN = pd.read_pickle(pkl_dir + "res_df_cotraining_CNN.pkl")
```

```python
fig, ax = plt.subplots(figsize=(7, 5))
sns.lineplot(data=res_df_only_HC_RF, x='threshold', y='f1_score', label='RandomForest (with HC features)', ax=ax, linewidth=2)
sns.lineplot(data=res_df_only_HC_LR, x='threshold', y='f1_score', label='LogisticRegression (with HC features)', ax=ax, linewidth=2)
sns.lineplot(data=res_df_only_HC_SVM, x='threshold', y='f1_score', label='SVM (with HC features)', ax=ax, linewidth=2)
sns.lineplot(data=res_df_only_ResNet, x='threshold', y='f1_score', label='ResNet18 (with CNN features)', ax=ax, linewidth=2)
sns.lineplot(data=res_df_only_VGG, x='threshold', y='f1_score', label='VGG16 (with CNN features)', ax=ax, linewidth=2)
sns.lineplot(data=res_df_only_MobileNet, x='threshold', y='f1_score', label='MobileNetV2 (with CNN features)', ax=ax, linewidth=2)
sns.lineplot(data=res_df_cotraining_HC, x='threshold', y='f1_score', label='Co-training (with HC features)', ax=ax, linewidth=2)
sns.lineplot(data=res_df_cotraining_CNN, x='threshold', y='f1_score', label='Co-training (with CNN features)', ax=ax, linewidth=2)
plt.legend(loc=(1.04, 0.365))
plt.ylabel("F1-Score")
plt.xlabel(r'Minimum Sv value (dB re 1 $m^{-1}$)')
plt.savefig('figures/min_Sv.png', dpi=300, bbox_inches='tight')
```

## Step 2. Sample size impact

```python
res_df_only_HC_sample = pd.read_pickle(pkl_dir + "res_df_only_HC_sample_size_5.pkl")
res_df_only_CNN_sample = pd.read_pickle(pkl_dir + "res_df_only_CNN_sample_size_5.pkl")
res_df_cotraining_HC_sample = pd.read_pickle(pkl_dir + "res_df_cotraining_HC_sample_size_5.pkl")
res_df_cotraining_CNN_sample = pd.read_pickle(pkl_dir + "res_df_cotraining_CNN_sample_size_5.pkl")
```

```python
fig, ax = plt.subplots(figsize=(7, 5))
sns.lineplot(data=res_df_only_HC_sample, x='labeled_num', y='f1_score', label='RandomForest (with HC features)', ax=ax, linewidth=2)
sns.lineplot(data=res_df_only_CNN_sample, x='labeled_num', y='f1_score', label='ResNet18 (with CNN features)', ax=ax, linewidth=2)
sns.lineplot(data=res_df_cotraining_HC_sample, x='labeled_num', y='f1_score', label='Co-training (with HC features)', ax=ax, linewidth=2)
sns.lineplot(data=res_df_cotraining_CNN_sample, x='labeled_num', y='f1_score', label='Co-training (with CNN features)', ax=ax, linewidth=2)
plt.xlabel('Number of labeled data')
plt.ylabel('F1-Score')
plt.ylim([0.5, 1.0])
plt.legend(loc=(1.04, 0.67))
plt.savefig('figures/sample_size.png', dpi=300, bbox_inches='tight')
```
